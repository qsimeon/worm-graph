```python
import pandas as pd
```

```python
df = pd.read_csv('neuron_names.csv')
```

```python
neuron_names = list(df.Neuron)
```

```python
neuron_names
```

```python
# open file in write mode
with open(r'neuron_names.txt', 'w') as fp:
    for neuron in neuron_names:
        # write each name on a new line
        fp.write("%s\n" % neuron)
    print('Done')
```
